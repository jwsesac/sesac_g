```python
df_NE = pd.read_excel('notExercise.xlsx',index_col='분류')
#파일설정
```

```python
df_jender = df_NE[1:3].T[1:6] #df 설정
df_age = df_NE[3:9].T[1:6]
df_student = df_NE[9:13].T[1:6]
```

```python
want_df = df_age #원하는 파이그래프 그리게 하기
title_list = list(want_df.index)
jpg_name = list(want_df.T["대분류"])[0]
for i in range(1,len(title_list)):
    plt.figure(figsize=(10,10))
    want_pie = want_df.loc[title_list[i]].plot.pie(figsize=(5,5),autopct='%.1f%%',counterclock=False,shadow=True)
    want_pie.set_ylabel("")
    want_pie.set_title(title_list[i])
    plt.savefig('save_{0}_fig{1}.png'.format(jpg_name,i),dpi=200)
    plt.show()
```
